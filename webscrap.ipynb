```python
import requests
from bs4 import BeautifulSoup
import json
import os

# URL de la página de aumentos
URL = 'https://tactics.tools/augments'
page = requests.get(URL)

# Analizar el contenido de la página
soup = BeautifulSoup(page.content, 'html.parser')

# Crear una lista para almacenar la información de los aumentos
augments_data = []

# Extraer la información de cada div que contiene el aumento
for augment in soup.find_all('div', class_=['flex', 'items-center']):
    # Extraer el nombre del aumento desde el div con clase "truncate"
    try:
        augment_name = augment.find('div', class_='truncate').get_text().strip()

        # Extraer la URL de la imagen desde la etiqueta img
        augment_image_url = augment.find('img')['src']

        # Agregar la información extraída a la lista
        augments_data.append({
            'name': augment_name,
            'image_url': augment_image_url
        })
    except:
        pass

# Guardar la información en un archivo JSON
with open('augments_data.json', 'w') as f:
    json.dump(augments_data, f, indent=4)

# Descargar las imágenes en una carpeta local
if not os.path.exists('augment_images'):
    os.makedirs('augment_images')

for augment in augments_data:
    img_data = requests.get(augment['image_url']).content
    with open(f'augment_images/{augment["name"].replace(" ", "_")}.jpg', 'wb') as img_file:
        img_file.write(img_data)

print("Web scraping completado y datos guardados.")
```

```python
import json

# Cargar los datos del archivo JSON
with open('augments_data.json', 'r') as file:
    augments = json.load(file)

# Diccionario para almacenar los aumentos únicos
unique_augments = {}

# Procesar los aumentos
for augment in augments:
    # Crear un nombre de archivo basado en el nombre del aumento
    augment_name = augment['name'].replace(" ", "_") + ".jpg"
    
    # Si el aumento no está ya en el diccionario, lo añadimos con el nuevo image_url
    if augment['name'] not in unique_augments:
        unique_augments[augment['name']] = {
            'name': augment['name'],
            'image_url': f'augment_images/{augment_name}'
        }

# Convertir el diccionario de aumentos únicos en una lista
unique_augments_list = list(unique_augments.values())

# Guardar los datos filtrados en un nuevo archivo JSON
with open('unique_augments_data.json', 'w') as output_file:
    json.dump(unique_augments_list, output_file, indent=4)

print("Aumentos filtrados y guardados con éxito.")
```
